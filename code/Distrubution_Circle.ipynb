This code uses altair to create a donut chart.

```python
import pandas as pd
import altair as alt
def urban_theme():
    markColor = "#A1674A"
    axisColor = "#000000"
    backgroundColor = "#fffeda"
    font = "Futura"
    labelFont = "Futura"
    sourceFont = "Futura"
    gridColor = "#CDCCCC"
    main_palette = ["#5D8ED4",
                    "#F6BE00",
                    "#7CDC97",
                    "#F1849A",
                    "#DC9FFF"
                   ]
    sequential_palette = [
                        "#FFFFFF"
                         ]
    return {
        "width": 700,
        "height": 380,   
        "cornerRadius": 90,
#        "autosize": "fit",
        "config": {
            "title": {
                "anchor": "middle",
                "fontSize": 24,
                "font": font,
                "fontColor": "#000000"
            },
            "axisX": {
               "domain": True,
               "domainColor": axisColor,
               "domainWidth": 1,
               "grid": False,
               "labelFontSize": 12,
               "labelFont": labelFont,
               "labelAngle": 0,
               "tickColor": "#FFFFFF",
               "tickSize": 5,
               "titleFontSize": 12,
               "titlePadding": 10,
               "titleFont": font,
               "title": "",
           },
           "axisY": {
               "domain": False,
               "grid": True,
               "gridColor": "#FFFFFF",
               "gridWidth": 1,
               "labelFontSize": 12,
               "labelFont": labelFont,
               "labelPadding": 8,
               "ticks": False,
               "titleFontSize": 12,
               "titlePadding": 10,
               "titleFont": font,
               "titleAngle": 0,
               "titleY": -10,
               "titleX": 18,
           },
           "background": backgroundColor,
           "legend": {
               "labelFontSize": 12,
               "labelFont": labelFont,
               "symbolSize": 100,
               "symbolType": "square",
               "titleFontSize": 12,
               "titlePadding": 10,
               "titleFont": font,
               "title": "Classes",
               "orient": "left",
               "offset": -30,
           },
           "view": {
               "stroke": "transparent",
           },
           "range": {
               "category": main_palette,
               "diverging": sequential_palette,
           },
           "area": {
               "fill": "#FFFFFF",
           },
           "line": {
               "color": markColor,
               "stroke": markColor,
               "strokewidth": 10,
           },
           "trail": {
               "color": markColor,
               "stroke": markColor,
               "strokeWidth": 0,
               "size": 1,
           },
           "path": {
               "stroke": markColor,
               "strokeWidth": 10,
           },
           "point": {
               "filled": True,
           },
           "text": {
               "font": sourceFont,
               "color": "#FFFFF",
               "fontSize": 11,
               "align": "middle",
               "fontWeight": 300,
               "size": 11,
               "dx": -30,
           }, 
       },
    }
    
    
import altair as alt
alt.themes.register("my_custom_theme", urban_theme)
alt.themes.enable("my_custom_theme")
```

```python
tempDict = {
    "Classes": ['MTH590', 'ENG300', 'PHY552', 'SPA300', 'CSC630',],
    "Hours/Week": [8,7,10,5,9]
}
temp = pd.DataFrame(tempDict)
```

```python
base = alt.Chart(temp, title="Hours Per Week I Spend on my Classes (not accurate at all)").encode(
    theta=alt.Theta("Hours/Week:Q", stack=True),
    radius=alt.Radius("Hours/Week:Q", scale=alt.Scale(type="sqrt", zero=True, rangeMin=10)),
    color=alt.Color("Classes:N")
).properties(
    width=400,
    height=500 
)
c1 = base.mark_arc(innerRadius=50, stroke="#FFFFFF", width=3000)
c2 = base.mark_text(
    radiusOffset=40, 
    fontSize=14).encode(text="Classes:N")
c1+c2
```
