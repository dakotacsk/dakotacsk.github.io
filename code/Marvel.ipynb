```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("/Users/dcoder/Desktop/school/csc630/network/archive (5)/edges.csv")
```

```python
common_comics = df['comic'].value_counts()[:5].index.tolist()

def isCommonComics(x):
    if x in common_comics:
        return True
    return False

df['isCommonComic'] = df['comic'].apply(isCommonComics)
```

```python
common_heros = df['hero'].value_counts()[:10].index.tolist()

def isCommonHeros(x):
    if x in common_heros:
        return True
    return False

df['isCommonHero'] = df['hero'].apply(isCommonHeros)
```

```python
df_common = df[df['isCommonComic']==True]
df_common = df_common[df_common['isCommonHero']==True]
```

```python
df_common
```

```python
df_common.reset_index(level=0, inplace=True)
```

```python
df_common.drop(["index","isCommonComic", "isCommonHero"], axis=1, inplace=True)
```

```python
df_common
```

```python
df_common.to_csv('marvel_hero_connections.csv')
```

```python
G = nx.from_pandas_edgelist(df_common, 
                            source='hero', 
                            target='comic', )
print("No of unique characters:", len(G.nodes))
print("No of connections:", len(G.edges))

from pyvis.network import Network
# create vis network
net = Network(notebook=True, width=1000, height=600)
# load the networkx graph
net.from_nx(G)
# show
net.show("example.html")
```
